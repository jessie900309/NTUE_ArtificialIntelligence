```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1)

# X 是 樣本點(二維)
# y 是 每個樣本被分到的群(centers=4 => 0、1、2、3)

# make_blobs() 參數
# n_samples 樣本數，default為100
# n_features 樣本為度，default為2
# centers 中心點的數量，default為3
# cluster_std 數據集的標準差，default為1.0
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1])
```

```python
print("X 樣本點 :", "\n", X)
print("y 分群結果 :", "\n", y)
```
