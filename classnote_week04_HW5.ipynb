```python
# 產生隨機樣本(.csv)
import numpy as np
import pandas as pd

# n_dots = 200
# X = np.random.rand(n_dots, 2) # 二維陣列
# print(X)
# (不好看，改用make_blobs)

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1)

df = pd.DataFrame(X, columns=['A', 'B'])
df.to_csv('aaa20231003_hw5.csv')
```

```python
# 讀取資料
csv_data = pd.read_csv('aaa20231003_hw5.csv')
csv_X = csv_data[["A", "B"]]
print(csv_X)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 3))
plt.scatter(csv_X['A'], csv_X['B'])

from sklearn.cluster import KMeans

k_means = KMeans(n_clusters=4)
k_means.fit(csv_X) # 訓練

y_kmeans = k_means.predict(csv_X) # 預測

plt.figure(figsize=(4, 3))
plt.scatter(csv_X['A'], csv_X['B'], c=y_kmeans)
```

```python
x0 = [7]
x1 = [-8] 
new_data = pd.DataFrame(list(zip(x0, x1)), columns =['A', 'B'])
new_data_pre = k_means.predict(new_data)
print("new_data : \n{}".format(new_data), "\n","predict : {}".format(new_data_pre))
```
