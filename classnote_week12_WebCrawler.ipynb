```python
import requests
# import bs4
from bs4 import BeautifulSoup

print(requests.__version__)
# print(bs4.__version__)
```

```python
web_html = requests.get("https://www.ntue.edu.tw/")
# print(web_html) # <Response [200]>
# print(web_html.text) # html doc

web_soup = BeautifulSoup(web_html.text, "html.parser")
print(web_soup)
```

```python
tag_title = web_soup.title
print(tag_title) # <title>國立臺北教育大學 </title>
print(tag_title.text) # 國立臺北教育大學
```

```python
tag_list = web_soup.find_all('h2', class_="mt-title")
# attrs = {attribute_key: attribute_value}
# class是python保留字，不能直接用
for tag in tag_list:
    print(tag.string)
```

```python
web_soup.select("span", limit=5)
# [<span>標題</span>, <span>副標題</span>]
```
