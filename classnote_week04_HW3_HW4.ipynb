```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.cluster import KMeans

k_means = KMeans(n_clusters=4)
k_means.fit(X) # 

y_kmeans = k_means.predict(X) # 

plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans)
```

```python
# 

# Datasets
plt.subplot(1,2,1)
plt.scatter(X[:, 0], X[:, 1])
plt.title('Datasets')

# Clusters
plt.subplot(1,2,2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans)
plt.title('Clusters')

plt.show()
```

```python
new_data = [[-4, -5]]
new_data_pre = k_means.predict(new_data)
print("new_data : {}".format(new_data), "\n","predict : {}".format(new_data_pre))
```
