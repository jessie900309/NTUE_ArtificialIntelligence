```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_dots = 20

# 利用 numpy.linspace() 在一定範圍內來均勻地撒點
# 比如說我們現在想要在 1 到 3 之間產生 5 個點 => numpy.linspace(1, 3, 5)
x = np.linspace(0, 1, n_dots)
# 設 y 為輸入 x 的每個元素都取平方根 => numpy.sqrt()
# 模擬雜訊擾動 + 0.2*np.random.rand(n_dots) - 0.1 
y = np.sqrt(x) + 0.2*np.random.rand(n_dots) - 0.1

# 利用 numpy.polyfit(x, y, deg) 進行多項式擬和
# x, y 表示擬和的座標點
# deg可指定擬和的多項式最高次方
def plot_polynomial_fit(x, y, order):
    p = np.poly1d(np.polyfit(x, y, order))
    t = np.linspace(0, 1, 200)
    plt.plot(x, y, 'ro', t, p(t), '-', t, np.sqrt(t), 'r--')
    return p
```

```python
plt.figure(figsize=(24, 5))

titles = ['Undeer Fitting', 'Fitting', 'Over Fitting']
models = [None, None, None]

# enumerate(sequence, [start=0])
# sequence:可迭代的物件  start:下标起始位置的值
# enumerate([1, 3, 10]) => [(0, 1), (1, 3), (2, 10)]
# enumerate([1, 3, 10], start=1) => [(1, 1), (2, 3), (3, 10)]
for index, order in enumerate([1, 3, 10]):
    plt.subplot(1, 3, index+1)
    models[index] = plot_polynomial_fit(x, y, order)
    plt.title(titles[index], fontsize=16)
```
